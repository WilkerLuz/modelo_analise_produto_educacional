Além disso, essas metricas podem nos fornecer valiosas informações, sobre o grupo de alunos, por exemplo:

 - Experiência na Área de Programação: Os campos relacionados à experiência em programação podem ajudar a identificar o nível de conhecimento prévio dos alunos, e assim, nos ajuda a entender o ritmo dos alunos e os requisitos de aprendizado.

 - Motivação para Escolher o Curso: Os motivos para escolher o curso de Bacharelado em Ciência da Computação fornecem informações sobre o que atraiu os alunos para a área, e desta maneira, nos ajuda a adaptar o ensino e os recursos para atender às expectativas e interesses dos alunos.

 - Tempo Dedicado aos Estudos: O número de horas que os alunos podem dedicar aos estudos fora das aulas pode afetar seu desempenho acadêmico. Desse modo, os alunos que podem dedicar mais tempo conseguem ter uma vantagem em relação aos que têm menos tempo disponível.


```python
#Importação das bibliotecas e leitura do arquivo .csv gerado após a manipulação da planilha

import pandas as pd
import matplotlib.pyplot as plt

df_perfil_alunos = pd.read_csv('perfil_alunos_PE.csv')
```

```python
df_perfil_alunos.describe()
```

```python
df_perfil_alunos.columns
```

```python


# Extraia os dados
faixa_etaria = df_perfil_alunos['faixa_etaria'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(faixa_etaria, labels=faixa_etaria.index, autopct='%1.1f%%', startangle=140)
plt.title('Faixa Etária dos Alunos')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```

```python
# Extraia os dados
genero = df_perfil_alunos['genero'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(genero, labels=genero.index, autopct='%1.1f%%', startangle=140)
plt.title('Gênero')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```

```python
# Extraia os dados
sabe_programar = df_perfil_alunos['experiencia_programacao'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(sabe_programar, labels=sabe_programar.index, autopct='%1.1f%%', startangle=140)
plt.title('Experiência com Programação')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```

```python
# Extraia os dados
pri_vez = df_perfil_alunos['primeira_vez'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(pri_vez, labels=pri_vez.index, autopct='%1.1f%%', startangle=140)
plt.title('Primeira vez na Disciplina?')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```

```python
# Extraia os dados
motivo = df_perfil_alunos['motivo'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(motivo, labels=motivo.index, autopct='%1.1f%%', startangle=140)
plt.title('Motivação para Fazer o Curso')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```

```python
# Extraia os dados
motivo_if = df_perfil_alunos['horas_dedicacao'].value_counts()

# Configurações do gráfico
plt.figure(figsize=(8, 6))
plt.pie(motivo_if, labels=motivo_if.index, autopct='%1.1f%%', startangle=140)
plt.title('Tempo de Dedicação aos Estudos')
plt.axis('equal')  # Assegura que o gráfico é desenhado como um círculo

# Mostra o gráfico
plt.show()
```
