```python
import dash
from dash import dcc, html, dash_table

import pandas as pd

from dash.dependencies import Input, Output

df_habilidades_previas = pd.read_csv('conhecimentos_previos_com_nota.csv')

dados = {'Aluno': df_habilidades_previas['Nome:'],
        'grupo': df_habilidades_previas['Q1']}

df_conhecimento = pd.DataFrame(dados)

# Crie um aplicativo Dash
app = dash.Dash(__name__)

# Layout do aplicativo
app.layout = html.Div([
    html.H1("Agrupamento de Alunos"),
    
    dcc.Dropdown(
        id='dropdown-q1',
        options=[
            {'label': grupo, 'value': grupo}
            for grupo in df_conhecimento['grupo'].unique()
        ],
        value=df_conhecimento['grupo'].iloc[0],
    ),
    
    dash_table.DataTable(
        id='table-alunos',
        columns=[
            {'name': 'Aluno', 'id': 'Aluno'},
        ],
        style_table={'height': '300px', 'overflowY': 'auto'},
    ),
])

# Callback para atualizar a tabela com base no grupo selecionado
@app.callback(
    Output('table-alunos', 'data'),
    Input('dropdown-q1', 'value')
)
def update_table(selected_group):
    filtered_data = df_conhecimento[df_conhecimento['grupo'] == selected_group]
    table_data = filtered_data[['Aluno']].to_dict('records')
    return table_data

if __name__ == '__main__':
    app.run_server(debug=True)
```
