Selecionar apenas as questões da Atividade do Produto Educacional chamado Atividades Iniciais em C

```python
import pandas as pd
from datetime import datetime

# Função para converter timestamp Unix em formato normal
def timestamp_para_data(timestamp):
    return datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')

# Aplicar a função para criar uma nova coluna com as datas no formato normal
df_vpl_atividades['DATA_ENTREGA'] = df_vpl_atividades['DUEDATE'].apply(timestamp_para_data)

# Salvar o DataFrame em um arquivo CSV
df_vpl_atividades.to_csv('vpl_atividades.csv', index=False)


# Exibir o DataFrame resultante
df_vpl_atividades
```
