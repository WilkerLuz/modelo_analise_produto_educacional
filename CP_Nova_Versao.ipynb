```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Dados fornecidos
data2 = {
    "NOME": df2['Nome:'],
    "Q1": df2['Q1_'],
    "Q2": df2['Q2_'],
    "Q3": df2['Q3_'], 
    "Q4": df2['Q4_'],
    "Q5": df2['Q5_'],
    "Q6": df2['Q6_'],
    "Q7": df2['Q7_'],
    "Q8": df2['Q8_'],
    "Q9": df2['Q9_'],
}

df = pd.DataFrame(data2)

# Remover colunas não necessárias para a clusterização
X = df.drop(columns=["NOME"])

# Determinando o número ideal de clusters usando o método do cotovelo
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plotando o método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inércia')
plt.title('Método do Cotovelo para determinar o número de clusters')
plt.show()

# Aplicando KMeans com o número de clusters ideal (assumindo 3 para este exemplo)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Mostrar resultados
df.to_csv('CP_Primeira_2.csv', index=False)

df
```

```python
import pandas as pd


df = pd.DataFrame(data2)

# Passo 2: Iterar sobre as colunas Q1_ até Q9_
for i in range(1, 10):
    col_name = f'Q{i}'
    new_col_name = f'{col_name}_nota'
    df[new_col_name] = df[col_name].replace({1: 1.2, 0: 0.6, -1: 0})

# Passo 3: Criar uma coluna com o total da soma de Q1_ até Q9_
df['Total_Soma'] = df[[f'Q{i}_nota' for i in range(1, 10)]].sum(axis=1)

# Mostrar resultados
df.to_csv('CP_primeira_3.csv', index=False)
df
```

```python
import pandas as pd

df_CP = pd.read_csv('CP_primeira_3.csv')

df_CP
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.cluster import KMeans

# Selecione as características desejadas
caracteristicas = ['Q1_nota', 'Q2_nota', 'Q3_nota', 'Q4_nota', 'Q5_nota', 'Q6_nota', 'Q7_nota', 'Q8_nota', 'Q9_nota', 'Total_Soma']

# Selecionar apenas as características do DataFrame
X = df_CP[caracteristicas]

# Escolha o número de clusters (n_clusters)
n_clusters = 3  # Altere o número de clusters conforme necessário

# Criar um modelo K-Means com o número desejado de clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Ajustar o modelo aos dados
kmeans.fit(X)

# Adicionar as informações do cluster de volta ao DataFrame
df_CP['Cluster'] = kmeans.labels_

# Exibir informações sobre os clusters
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=caracteristicas)
cluster_info = df_CP.groupby('Cluster').size()
print("Centros dos Clusters:")
print(cluster_centers)
print("\nContagem de amostras em cada cluster:")
print(cluster_info)

# Preparar os dados para o gráfico de barras empilhadas
fig = go.Figure()

# Cores para os clusters
cluster_colors = {0: 'red', 1: 'blue', 2: 'green'}  # Ajuste conforme o número de clusters

for cluster in range(n_clusters):
    cluster_data = df_CP[df_CP['Cluster'] == cluster]
    for questao in ['Q1_nota', 'Q2_nota', 'Q3_nota', 'Q4_nota', 'Q5_nota', 'Q6_nota', 'Q7_nota', 'Q8_nota', 'Q9_nota']:
        fig.add_trace(go.Bar(
            x=cluster_data['NOME'],
            y=cluster_data[questao],
            name=f'{questao} (Cluster {cluster})',
            marker_color=cluster_colors[cluster]
        ))

# Atualizar layout do gráfico
fig.update_layout(
    barmode='stack',
    title='Respostas dos Alunos por Questão com Clusters',
    xaxis=dict(title='Nome dos Alunos'),
    yaxis=dict(title='Notas'),
    legend=dict(title='Questões e Clusters')
)

# Exibir o gráfico
fig.show()

```
