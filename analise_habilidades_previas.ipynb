```python
# Importando as bibliotecas pandas, matplolib e numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Iportando o Arquivo CSV para a análise dos dados, 
#Abaixo será gerado um DataFrame para a manipulação usando o Pandas

df_habilidades_previas = pd.read_csv('conhecimentos_previos.csv')
```

```python
#Apresentação do DataFrame

df_habilidades_previas.head()
```

```python
"""Geraçao de um gráfico de barras usando matplotlib para o agrupamento dos conhecimentos prévios"
dos alunos nas questões do teste. """

# Contar quantas vezes cada nota aparece
contagem_notas = df_habilidades_previas['NOTA_Q1'].value_counts()

# Plotar o gráfico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(contagem_notas.index, contagem_notas.values)
plt.bar(contagem_notas.index, contagem_notas.values)

plt.xlabel('Conhecimento Prévio')
plt.ylabel('Número de Alunos')
plt.title('Habilidades Questão 01')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom')

plt.show()
```

```python
"""Listagem dos anos por grupo, sabendo que aquele no Grupo N(não), não possuem conhecimento prévio, 
                                                    Grupo P(Parcialmente), possui algum conhecimento, e 
                                                    Grupo S(Sim), possui o conhecimento prévio necessário
"""
# Agrupamento com base na coluna 'Grupo'
grupos = df_habilidades_previas.groupby('NOTA_Q1')

# Listar os elementos de cada grupo
for grupo, elementos in grupos:
    print(f'Grupo {grupo}:')
    print(elementos['ID'])
    print()
```

```python

# Defina as colunas de questões (excluindo a coluna 'aluno')
colunas_questoes = ['NOTA_Q1', 'NOTA_Q2', 'NOTA_Q3','NOTA_Q4','NOTA_Q5','NOTA_Q6','NOTA_Q7','NOTA_Q8','NOTA_Q9' ]

# Crie um novo DataFrame para armazenar as contagens
contagem_respostas = pd.DataFrame()

# Itere sobre as colunas de questões e calcule a contagem para cada uma
for coluna in colunas_questoes:
    contagem = df_habilidades_previas[coluna].value_counts()
    contagem_respostas[coluna] = contagem

# Adicione a coluna 'aluno' ao novo DataFrame
contagem_respostas['ID'] = df_habilidades_previas['ID']

# Exiba o novo DataFrame com as contagens
print(contagem_respostas)
```

```python
# Crie uma tabela de contingência (cross-tab) entre as duas colunas
tabela_contingencia = pd.crosstab(df_habilidades_previas['Você possui alguma experiência na Área de Programação?'],
                                  df_habilidades_previas['NOTA_Q5'])

# Plote o gráfico de barras empilhadas
tabela_contingencia.plot(kind='bar', stacked=True, figsize=(8, 6))

plt.xlabel('Experiência')
plt.ylabel('Contagem')
plt.title('Relação entre Experiência e Habilidade')
plt.legend(title='Habilidade')
plt.show()
```
