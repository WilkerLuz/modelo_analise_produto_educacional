```python
# Importando matplotlib, pandas e numpy
import pandas as pd
import matplotlib.pyplot as plt


df_perfil_alunos = pd.read_csv('perfil_alunos.csv')
```

```python
df_perfil_alunos.head()
```

```python
# Agrupamento por faixa etária

agrup_faixa_etaria = df_perfil_alunos.groupby('Qual sua faixa etária (idade)?').size()
```

```python
agrup_faixa_etaria
```

```python
#Criando o gráfico da distribuição de Faixa Etária da Turma

plt.figure(figsize=(6, 6))  # Ajuste o tamanho da figura conforme necessário
cores=['#FFE4E1', '#EEE8AA', '#E6E6FA', '#B0E0E6',]
plt.pie(agrup_faixa_etaria, labels=agrup_faixa_etaria.index, colors=cores, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Isso garante que o gráfico seja uma pizza perfeita (círculo)
plt.title('Distribuição de Faixa Etária na Turma')
plt.show()
```

```python
import numpy as np
quant_faixa = np.array([3, 31, 1, 6])
idade = ['Acima de 30 anos','Entre 18 e 25 anos','Entre 26 e 30 anos','Menor de 18 anos']

cores=['#FFE4E1', '#EEE8AA', '#E6E6FA', '#B0E0E6',]
# o atributo explode indica que fatia do gráfico será destacada. No exemplo abaixo, será a primeira fatia. A quantidade de valores é igual ao número de fatias do gráfico. 
explode = (0, 0, 0, 0)  # explode 1st slice

# Atribuindo um título ao gráfico
plt.title('Faixa Etária da Turma')

plt.pie(quant_faixa, explode=explode, labels=idade, colors=cores, autopct='%1.1f%%', shadow=True, startangle=90)

#Adiciona Legenda
plt.legend(idade, bbox_to_anchor=(1.3, 1.3),loc='upper right')

#Centraliza o gráfico
plt.axis('equal')

#Ajusta o espaçamento para evitar o recorte do rótulo
plt.tight_layout()

plt.show()
```
